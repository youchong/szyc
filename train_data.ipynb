```python
train=pd.read_csv('data/july_aug.csv')
test=pd.read_csv('data/test_id_Aug_agg_private5k.csv')
poi=pd.read_csv('data/pio_treated.csv')
weather=pd.read_csv('data/weather_treated.csv')
```

```python
weather['hour']=weather['time'].str.split(':',expand=True)[0]
```

```python
weather=weather.groupby(['date','hour']).last().reset_index()
```

```python
gb=train.groupby(['create_date','create_hour','start_geo_id','end_geo_id'])
```

```python
train=gb['end_geo_id'].agg({'count':np.size}).reset_index()
```

```python
train.shape
```

```python
def merge_weather(data):
    data['dt']=data['create_date']+data['create_hour'].astype(str)
    weather['dt']=weather['date']+weather['hour'].astype(str)
    data=data.merge(weather,how='left',on='dt')
    data.drop(['dt'],axis=1,inplace=True)
    return data
```

```python
train=merge_weather(train)
```

```python
train.shape
```

```python
test=merge_weather(test)
```

```python
def merge_poi(data):
    data=data.merge(poi,how='left',left_on='start_geo_id',right_on='geo_id')
    data=data.merge(poi,how='left',left_on='end_geo_id',right_on='geo_id')
    return data
```

```python
train=merge_poi(train)
```

```python
train.head()
```

```python
test=merge_poi(test)
```

```python
count=train['count']
train.drop(['count'],axis=1,inplace=True)
train.insert(0,'count',count)
```

```python
sgi=train['start_geo_id']
train.drop(['start_geo_id'],axis=1,inplace=True)
train.insert(1,'start_geo_id',sgi)
```

```python
egi=train['end_geo_id']
train.drop(['end_geo_id'],axis=1,inplace=True)
train.insert(2,'end_geo_id',egi)
```

```python
train.drop(['date','time','geo_id_x','geo_id_y'],axis=1,inplace=True)
```

```python
trd=train['create_date'].str.split('-',expand=True)
ted=test['create_date'].str.split('-',expand=True)
```

```python
train.insert(4,'year',trd[0].astype(int))
train.insert(5,'month',trd[1].astype(int))
train.insert(6,'day',trd[2].astype(int))
test.insert(4,'year',ted[0].astype(int))
test.insert(5,'month',ted[1].astype(int))
test.insert(6,'day',ted[2].astype(int))
```
