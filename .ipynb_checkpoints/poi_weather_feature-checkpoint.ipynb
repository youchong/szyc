```python
weather=pd.read_csv('data/weather.csv')
```

```python
weather.head()
```

```python
weather.tail()
```

```python
weather['text'].value_counts()
```

```python
weather['text'].unique().tolist()
```

```python
weather['text'].replace(['晴', '多云', '阴', '小雨', '雷阵雨', '中雨', '大雨', '阵雨'],[2,1,6,5,3,4,1,2],inplace=True)
```

```python
weather['wind_direction'].value_counts()
```

```python
weather['wind_direction'].unique().tolist()
```

```python
weather['wind_direction'].replace(['南', '西南', '东南', '西', '西北', '东', '东北', '北'],[5,8,4,3,1,7,6,2],inplace=True)
```

```python
weather.head()
```

```python
weather[['date','time']]=weather['date'].str.split(' ',expand=True)
```

```python
weather.head()
```

```python
time=weather.pop('time')
weather.insert(1,'time',time)
```

```python
weather.head()
```

```python
weather.to_csv('data/weather_treated.csv',index=False)
```
